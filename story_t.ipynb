### Versão Simples

```python
df = pd.DataFrame(data = {'manuf':[8.2,8.6,8.8,10.0,10.3,11,11.6,11.6,12.3,12.9],
                          'text':["Engine power is less than expected",
                                  "Tires make excessive noise while driving",
                                  "Engine makes abnormal/excessive noise",
                                  "Seat material concerns",
                                  "Excessive wind noise",
                                  "Hesitation or delay when shifting",
                                  "Bluetooth system has poor sound quality",
                                  "Steering system/wheel has too much play",
                                  "Bluetooth system is difficult to use",
                                  "Front seat navigation controls"
                                  ]
                          })
df
```

```python
def set_label_y_in_bar(bars):
    for bar in bars:
        width = bar.get_width()
        label_x_pos = width - 1  # ajuste a posição do rótulo
        label_text = f"{width:.1f}"  # formatando o texto do rótulo
        ax.text(label_x_pos, bar.get_y() + bar.get_height() / 2, label_text,
                ha='left', va='center', fontsize=8, color='white')
```

```python
lista_cores = ['#524e4e','#524e4e','#524e4e','#524e4e','#524e4e',
               '#a32e2e','#524e4e','#a32e2e','#a32e2e','#524e4e']
x = np.array(range(1, df.shape[0]+1))
fig ,ax = plt.subplots(figsize = (6,4))

graf = ax.barh(df.text,df.manuf,color = lista_cores)
## Título Superior
ax.text(x = -6,
        y = 10.5,
        s = "Car manufacturer Z: top 10 design concerns",
        fontsize = 10,
        color = "#000000",
        weight = 'bold'
)
## Título Inferior
ax.text(x = -2.3,
        y = 9.8,
        s = "design issue | concerns per 1000",
        fontsize = 8,
        color = "#747474", #cinza
        weight = 'light'
)

# Texto 1


ax.text(x = 13.5,
        y = 8,
        s = "Comments indicate that\n" +
        "noisy tire issues are most\n"+
        "apparent $\\bf{in}$ $\\bf{the}$ $\\bf{rain}.$" ,
        fontsize = 10,
        color = "black"
)
# Texto 2
ax.text(x = 13.5,
        y = 5.2,
        s = "Complaints about engine\n" +
        "noise were commonly cited\n"+
        "$\\bf{after}$ $\\bf{the}$ $\\bf{car}$ $\\bf{had}$" +
        " $\\bf{not}$ $\\bf{been}$\n$\\bf{driven}$ $\\bf{for}$" +
        "$\\bf{a}$ $\\bf{while}.$ "  ,
        fontsize = 10,
        color = "black"
)
# Texto 3
ax.text(x = 13.5,
        y = 3,
        s = "Excessive wind noise is\n" +
        "noted primarily for $\\bf{freeway}$\n"+
        "$\\bf{driving}$ $\\bf{at}$ $\\bf{high}$ $\\bf{speeds}.$"  ,
        fontsize = 10,
        color = "black"
)
ax.tick_params(labelcolor='#524e4e', labelsize=8)

# Removendo traços do eixo y
ax.yaxis.set_ticks_position('none')  # remove os traços do eixo y


ax.spines['left'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)

set_label_y_in_bar(graf)
plt.gca().axes.get_xaxis().set_visible(False)
plt.show()
```

```python
df = pd.DataFrame(data = {'Year':[2001,2002,2003,2004,2005,2006,2007,2008,2009,2010],
                          'Television':[74,82,80,74,73,72,74,70,70,66],
                          'Newspaper':[45,42,50,46,36,36,34,35,32,31],
                          'Internet':[13,14,20,24,20,24,24,40,35,41],
                          'Radio':[18,21,18,21,16,14,13,18,17,16]})
df
```

```python
fig, ax = plt.subplots(figsize = (5,3))
ax.plot(df.Year,df.Television,color = "#747474")
ax.plot(df.Year,df.Newspaper,color = "#747474")
ax.plot(df.Year,df.Internet,color = "#3976b3",linewidth = 3)
ax.plot(df.Year,df.Radio,color = "#747474")

ax.scatter(2001,df.Television[0],color = "#747474")
ax.scatter(2001,df.Newspaper[0],color = "#747474")
ax.scatter(2001,df.Internet[0],color = "#3976b3")
ax.scatter(2001,df.Radio[0],color = "#747474")

ax.scatter(2010,df.Television[9],color = "#747474")
ax.scatter(2010,df.Newspaper[9],color = "#747474")
ax.scatter(2010,df.Internet[9],color = "#3976b3")
ax.scatter(2010,df.Radio[9],color = "#747474")

## Título Superior
ax.text(x = 1996,
        y = 100.5,
        s = "How people get their news",
        fontsize = 15,
        color = "#000000",
        weight = 'bold'
)
## Título Inferior
ax.text(x = 1996,
        y = 90.5,
        s = "An increasing proportion cite the",
        fontsize = 10,
        color = "#000000"
)
ax.text(x = 2002,
        y = 90.5,
        s = "internet",
        fontsize = 10,
        color = "#3976b3"
)

ax.text(x = 2003.5,
        y = 90.5,
        s = "as their primary news source",
        fontsize = 10,
        color = "#000000"
)



ax.spines['left'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

for i in df.columns:
  if i != 'Year':
    if i != 'Internet':
      ax.text(x = 1999.8 ,y = df.loc[0,i] - 1, s =f"{i}",horizontalalignment='right',color = "#747474")
      ax.text(x = 2000 ,y = df.loc[0,i] - 1, s =f"{df.loc[0,i] }%",color = "#747474")
      ax.text(x = 2010.5 ,y = df.loc[9,i] - 1, s =f"{df.loc[9,i] }%",color = "#747474")
    else:
      ax.text(x = 1999.8 ,y = df.loc[0,i] - 1, s =f"{i}",horizontalalignment='right',color = "#3976b3")
      ax.text(x = 2000 ,y = df.loc[0,i] - 1,  s =f"{df.loc[0,i] }%",color = "#3976b3")
      ax.text(x = 2010.5 ,y = df.loc[9,i] - 1, s =f"{df.loc[9,i] }%",color = "#3976b3")


plt.gca().axes.get_yaxis().set_visible(False)

plt.show()
```
